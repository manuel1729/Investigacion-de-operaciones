Creamos con add_edge("","",capacity=?) para unir los nodos y arcos

```python
#Gráica dirigida
G.add_edge("A","B",capacity=2)
G.add_edge("A","C",capacity=2)
G.add_edge("B","C",capacity=2)
G.add_edge("B","D",capacity=2)
G.add_edge("C","D",capacity=2)

#Para hacer una gráfica no dirigida
H = nx.Graph()
H.add_edge(1, 2)
e = (2, 3)
H.add_edge(*e)
H.add_edges_from([(1,7),(3,4),(4, 5),(5,6), (6, 7)])
```

Agregamos atributos a nodos y aristas con un valor en especifíco

```python
#atributos en nodos
A = nx.Graph()
A.add_edges_from([(1, 2), (2, 3), (1, 7), (3, 4), (4, 5), (5, 6), (6, 7)])

A.add_nodes_from([1, 2])
A.nodes[1]["rol"] = "MEDEL"
A.nodes[2]["rol"] = "OUMPALUMPA"

#atributos en aristas
A[1][2]["pesos"] = 17

print(A.nodes(data=True))
print(A.edges(data=True))
```

Uso del multigrafo imprimiendo con matplotlib

```python
import matplotlib.pyplot as plt
MG = nx.MultiGraph()

MG.add_nodes_from([ "A" , "B" , "C" ])
MG.add_edges_from([( "A" , "B" ), ( "B" , "C" )])
nx.draw_networkx(MG)
plt.show()
```

### Creación de muchos graficos

Creamos un gráfico de grado n del 1 al n

```python
M_8 = nx.complete_graph(8)
nx.draw_networkx(M_8)
```

Generamos una estrella de centro m y picos n (m,n)

```python
L = nx.complete_bipartite_graph(5, 7)
nx.draw_networkx(L)
```

Otra forma de gráfico

```python
ba= nx.barbell_graph(15, 10)
nx.draw_networkx(ba)
```

Gráfico lollipop

```python
lollipop = nx.lollipop_graph(10, 20)
nx.draw_networkx(lollipop)
```

Gráfico de Erdős-Rényi o gráfico binomial

```python
er = nx.erdos_renyi_graph(50, 0.20)
nx.draw_networkx(er)
```

Gráfico de Watts-Strogratz

```python
WS = nx.watts_strogatz_graph(20, 2, 0.2)
nx.draw_networkx(WS)
```

Gráfico aleatorio utilizando el vínculo preferencial de Barabási-Albert

```python
BA = nx.barabasi_albert_graph(60, 4)
nx.draw_networkx(BA)
```

Gráfico de lobster

```python
R = nx.random_lobster(90, 0.5, 0.5)
nx.draw_networkx(R)
```
